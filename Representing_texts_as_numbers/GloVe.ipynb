```python
# دالة مخصصة لتوليد القاموس
# مفردة/قائمة
def load_glove_embeddings(file_path):
    embeddings = {}
    # فتح الملف للقراءة
    with open(file_path, 'r', encoding='utf-8') as f:
        # الدوران على أسطر الملف
        for line in f:
            # تقسيم السطر وفق الفراغات
            values = line.split()
            # المفردة في الفهرس الأول
            word = values[0]
            # قائمة التضمين
            vector = [float(val) for val in values[1:]]
            # الإضافة للقاموس
            embeddings[word] = vector
    return embeddings
```

```python
# مسار الملف
glove_file_path = 'models/glove.6B.100d.txt'
# توليد قاموس التضمين
glove_embeddings = load_glove_embeddings(glove_file_path)
```

```python
# طول قوائم التضمين
embedding_dim=50
# دالة مخصصة لتوليد تضمين نص
def text_to_embeddings(text, embeddings):
    # تقسيم النص إلى مفردات
    words = text.lower().split()
    # توليد قائمة التضمين للنص
    embeddings_list = [embeddings.get(word, [0] * embedding_dim) for word in words]
    return embeddings_list
```

```python
# مثال
example_text = "This is a sample text."
example_embeddings = text_to_embeddings(example_text, glove_embeddings)
example_embeddings
```

```python
# مثال
word1 = "man"
word1_embeddings = text_to_embeddings(word1, glove_embeddings)
word2="woman"
word2_embeddings = text_to_embeddings(word2, glove_embeddings)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# حساب التجيب كمقياس تشابه
cosine = cosine_similarity(word1_embeddings, word2_embeddings)
print("Cosine Similarity : ")
print(cosine)
```
