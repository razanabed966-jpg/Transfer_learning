```python
documents = ["Learning Python programming is fast as Python is easy",
             "Python is great programming language"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

max_features=3
tfidf_vectorizer = TfidfVectorizer(max_features=max_features)
tfidf_vectors = tfidf_vectorizer.fit_transform(documents)
vectors = tfidf_vectorizer.fit_transform(documents)
print(vectors)
```

```python
feature_names = tfidf_vectorizer.get_feature_names_out()
print("Feature Names:")
print(feature_names)
```

```python
print(vectors.toarray())
```

```python
import pandas as pd

example_df=pd.DataFrame(data=tfidf_vectors.toarray(),columns=feature_names)
example_df
```
