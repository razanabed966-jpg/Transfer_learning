```python
from gensim.models import KeyedVectors
model_path = '/content/drive/MyDrive/Colab Notebooks/Hsoub/L06/GoogleNews-vectors-negative300.bin'
google_word2vec_model = KeyedVectors.load_word2vec_format(model_path, binary=True)
```

```python
google_word2vec_model.key_to_index
```

```python
vocab_size = len(google_word2vec_model.key_to_index)
print("Vocabulary size:", vocab_size)
```

```python
word= 'Python'
if word in google_word2vec_model:
    word_vector_python = google_word2vec_model[word]
    print("Vector for ", word, " :")
    print(len(word_vector_python))
    print(word_vector_python)
else:
    print(word, " is not in the vocabulary.")
```

```python
word= 'Python3'
if word in google_word2vec_model:
    word_vector_python = google_word2vec_model[word]
    print("Vector for ", word, " :")
    print(word_vector_python)
else:
    print(word, " is not in the vocabulary.")
```

```python
word1 = 'city'
word2 = 'town'
similarity_score = google_word2vec_model.similarity(word1, word2)
print(f"Similarity between '{word1}' and '{word2}': {similarity_score}")
```

```python
word1 = 'city'
word2 = 'boy'
similarity_score = google_word2vec_model.similarity(word1, word2)
print(f"Similarity between '{word1}' and '{word2}': {similarity_score}")
```

```python
word="Python"
similar_words = google_word2vec_model.most_similar(word)
print("Most similar words to ",word, " : ", similar_words)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

from nltk.tokenize import word_tokenize
sentence1 = "I love programming in Python"
sentence2 = "Python is a great programming language"
tokenized_sentence1 = word_tokenize(sentence1)
tokenized_sentence2 = word_tokenize(sentence2)

vectors_sentence1 = [google_word2vec_model[word] for word in tokenized_sentence1 if word in google_word2vec_model]
vectors_sentence2 = [google_word2vec_model[word] for word in tokenized_sentence2 if word in google_word2vec_model]

avg_vector_sentence1 = sum(vectors_sentence1) / len(vectors_sentence1)
avg_vector_sentence2 = sum(vectors_sentence2) / len(vectors_sentence2)

similarity_score = cosine_similarity([avg_vector_sentence1], [avg_vector_sentence2])

print(f"Similarity between '{sentence1}' \n and '{sentence2}':\n {similarity_score}")
```
